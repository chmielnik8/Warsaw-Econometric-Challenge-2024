```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv("/content/data_municipalities.csv")
df.isna().sum()
df.describe()

df = df.drop(df.columns[np.arange(58, 100, 1)], axis=1)
df = df.drop([value for value in df.columns if value.endswith('_m')], axis=1)
df = df.drop([value for value in df.columns if value.endswith('_f')], axis=1)
```

```python
for i, col in enumerate(df.columns[5:]):
    df[col[:-6] + "_percentage"] = df[col] / df['population_total']
df.columns[22:]
```

```python
df.sort_values(by='population_80_84_percentage', ascending=False)
```

```python
int(len(df)/10)
```

```python
best = df.nlargest(n=int(len(df)/10), columns='percent_vaccinated')
worst = df.nsmallest(n=int(len(df)/10), columns='percent_vaccinated')
```
